# Youtube_Scraper-API

Scraping de Youtube en passant par Youtube Data API v3 (Google Cloud).

```python
import requests
import pandas as pd
```

```python
API_KEY = "your_API_key"
```

```python
FOXNEWS_ID = "UCXIJgqnII2ZOINSWNOGFThA"
CNN_ID = "UCupvZG-5ko_eiXAupbDfxWw"
MSNOW_ID = "UCaXkIU1QidjPwiAYu6GcHjg"
```

```python

API_KEY = API_KEY
MOT_CLE = "Kamala Harris"
CHANNEL_ID = MSNOW_ID
DATE_DEBUT = "2024-08-12T00:00:00Z"
DATE_FIN = "2024-09-25T23:59:59Z"

resultats = []
next_page_token = None

while True:
    params = {
        'part': 'snippet',
        'channelId': CHANNEL_ID,
        'q': MOT_CLE,
        'type': 'video',
        'publishedAfter': DATE_DEBUT,
        'publishedBefore': DATE_FIN,
        'maxResults': 50,
        'pageToken': next_page_token,
        'key': API_KEY
    }

    response = requests.get("https://www.googleapis.com/youtube/v3/search", params=params)
    data = response.json()

    for item in data.get('items', []):
        resultats.append({
            'title': item['snippet']['title'],
            'channel': item['snippet']['channelTitle'],
            'upload_date': item['snippet']['publishedAt'],
            'url': f"https://www.youtube.com/watch?v={item['id']['videoId']}"
        })

    next_page_token = data.get('nextPageToken')
    if not next_page_token:
        break

df = pd.DataFrame(resultats)
print(df)
```
