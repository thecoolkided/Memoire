```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
fichiers_csv = {
    "Fox News": "/Users/Manon/Desktop/Fac/ENC/Mémoire/Computer/Scrapping/Datasets/Youtube/KeyW_immigation/FoxNews/foxnews_immigration_clean.csv",
    "CNN": "/Users/Manon/Desktop/Fac/ENC/Mémoire/Computer/Scrapping/Datasets/Youtube/KeyW_immigation/CNN/CNN_immigration_clean.csv",
    "MS NOW": "/Users/Manon/Desktop/Fac/ENC/Mémoire/Computer/Scrapping/Datasets/Youtube/KeyW_immigation/MSNOW/MSNOW_immigration.csv",
}

noms_datasets = []
totaux_videos = []

for nom, chemin in fichiers_csv.items():
    try:
        df_temp = pd.read_csv(chemin)
        noms_datasets.append(nom)
        totaux_videos.append(len(df_temp))
    except FileNotFoundError:
        print(f"Erreur : Le fichier '{chemin}' est introuvable.")

if totaux_videos:
    fig, ax = plt.subplots(figsize=(8, 5.5))

    couleurs = ["#2b5c8f", "#d95f02", "#7570b3"]
    bars = ax.bar(
        noms_datasets, totaux_videos, color=couleurs, width=0.55, zorder=3
    )

    ax.grid(axis="y", linestyle="--", alpha=0.7, zorder=0)

    for spine in ["top", "right", "left"]:
        ax.spines[spine].set_visible(False)
    ax.spines["bottom"].set_color("#cccccc")

    for bar in bars:
        height = bar.get_height()
        ax.annotate(
            f"{int(height):,}",
            xy=(bar.get_x() + bar.get_width() / 2, height),
            xytext=(0, 6),
            textcoords="offset points",
            ha="center",
            va="bottom",
            fontsize=11,
            fontweight="bold",
            color="#333333",
        )

    ax.set_title(
        "Comparaison du nombre total de vidéos mentionnant le mot-clé 'immigration'",
        fontsize=15,
        fontweight="bold",
        color="#1a1a1a",
        pad=20,
    )
    ax.set_xlabel(
        "Datasets", fontsize=11, fontweight="semibold", color="#444444", labelpad=12
    )
    ax.set_ylabel(
        "Nombre total de vidéos",
        fontsize=11,
        fontweight="semibold",
        color="#444444",
        labelpad=12,
    )

    plt.tight_layout()
    plt.show()
```
