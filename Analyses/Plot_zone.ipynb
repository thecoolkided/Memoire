# Plot_Zone

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_trump_x = pd.read_csv("/Users/Manon/Desktop/Fac/ENC/Mémoire/Computer/Scrapping/Datasets/Scraper_X/Trump/df_trump_x_clean.csv")
df_trump_truth = pd.read_csv("/Users/Manon/Desktop/Fac/ENC/Mémoire/Computer/Scrapping/Datasets/Trump_Truth/Trump_truth_clean_finale.csv")
df_harris_x = pd.read_csv("/Users/Manon/Desktop/Fac/ENC/Mémoire/Computer/Scrapping/Datasets/Scraper_X/Harris/df_harris_x_clean.csv")
```

## Plot_nombre total de posts

### Plot_1 : Trump_X vs Trump_Truth

```python
colors = ["black", "#6B3FA0"]

pd.Series({
    'X': len(df_trump_x),
    'Truth Social': len(df_trump_truth),
}).plot(kind='bar', title='Nombre de posts de Trump sur X et Truth Social de mi-août à fin septembre 2024', color=colors)
plt.xticks(rotation=0)
plt.show()
```

### Plot_2 : Trump_X vs Harris_X

```python
colors = ["#E9141D", "darkblue"]

pd.Series({
    'Trump_X': len(df_trump_x),
    'Harris_X': len(df_harris_x),
}).plot(kind='bar', title='Comparaison du nombre de tweets de Kamala Harris et de Donald Trump sur la période de mi-août à fin septembre 2024', color=colors)
plt.xticks(rotation=0)
plt.show()
```

### Plot_3 : Trump_total vs Harris_X

```python
df_trump_total = len(df_trump_x) + len(df_trump_truth)
df_trump_total
```

```python
colors = ["#E9141D", "darkblue"]

pd.Series({
    'Trump_total': df_trump_total,
    'Harris_X': len(df_harris_x),
}).plot(kind='bar', title='Comparaison du nombre de posts de Kamala Harris et de Donald Trump sur la période de mi-août à fin septembre 2024', color=colors)
plt.xticks(rotation=0)
plt.show()
```
